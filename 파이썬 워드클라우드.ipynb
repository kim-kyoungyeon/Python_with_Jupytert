```python
file = open('..data/naver_blog.txt', 'r', encoding='utf-8')
c = file.read()
c
```

```python
from konlpy.tag import *
okt = Okt()
```

이 클래스들은 다음과 같은 메서드를 공통적으로 제공한다.

* `nouns` : 명사 추출
* `morphs` : 형태소 추출
* `pos` : 품사 부착

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 한글 문제
# matplotlib의 기본 폰트에서 한글이 지원되지 않기 때문에
# matplotlib의 폰트 변경 필요
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
from nltk import Text

kolaw = Text(okt.nouns(c), name="kolaw")
kolaw.plot(50)
plt.show()
```

```python
from wordcloud import WordCloud

# 자신의 컴퓨터 환경에 맞는 한글 폰트 경로를 설정
font_path = 'C:/Windows/Fonts/HMKMRHD.TTF'

plt.figure(figsize=(10, 6)) 

wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path)
plt.imshow(wc.generate_from_frequencies(kolaw.vocab()))
plt.axis("off")
plt.show()
```

```python
# 불필요한 단어 제외

# stopword = ['서울시', '도시', '서울', '시민', '수', '코로나','대중교통','때문', '일반', '교통', '그', '것', '이용']

# 불필요한 단어를 파일로 저장한 경우
f = open('stopword.txt', 'r', encoding='utf-8')
stopword = f.read()
# stopword

kolaw = [i for i in kolaw if i not in stopword]
kolaw = Text(kolaw)
```
