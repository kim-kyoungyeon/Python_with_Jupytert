경우에 따라 데이터프레임에 인덱스로 들어가 있어야 할 데이터가
일반 데이터 열에 들어가 있거나
반대로 일반 데이터 열이어야 하는데 인덱스로 들어가 있는 경우가 존재
인덱스와 열을 교환하는 방법이 필요

데이터프레임의 인덱스 설정하고 제거하는 함수
set_index() : 기존의 행 인덱스를 제거하고 데이터 열 중에서 하나를 인덱스로 설정
              (열 -> 인덱스)
reset_index() : 기존의 행 인덱스를 제거하고 인덱스를 열로 추가
                (인덱스 -> 열)

```python
import pandas as pd
import numpy as np
```

```python
# 데이터프레임을 생성
# np.vstack() 함수를 사용해서 데이터프레임 생성
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
# np.vstack() 함수 : 2 개의 array를 위/아래로 합침
np.vstack([a,b])
```

```python
np.hstack([a,b])
```

```python
# np.vstack() 함수를 사용해서 데이터프레임 생성
df = pd.DataFrame(np.vstack([a,b]),columns=['C1','C2','C3'])
df
```

```python
# 인덱스 조작에 사용할 데이터프레임 생성 : np.stack() 사용
# 실행할 때마다 동일한 값으로 실행되도록 시드 값 설정
np.random.seed(2)
# 4행 5열의 데이터프레임 생성
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                            np.round(np.random.rand(3,5),2)]), index =['C1','C2','C3','C4'])
df1
```

```python
# 'C1', 'C2','C3', 'C4'를 열 인덱스 사용
df1.T
```

```python
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                            np.round(np.random.rand(3,5),2)]).T, columns =['C1','C2','C3','C4'])
df1
```

set_index()
- 기존의 행 인덱스를 제거하고
- 데이터 열 중에서 하나를 인덱스로 설정
- 열 -> 인덱스

```python
df2 = df1.set_index('C1')
df2
```

```python
# C2열을 인덱스로 지정
# 기존의 인덱스 C1은 없어짐
df2.set_index('C2')
```

reset_index() : 기존의 행 인덱스를 제거하고
- 인덱스를 열로 추가 (인덱스 -> 열)
- 이때 인덱스 열은 열 이름 맨 앞에 삽입 (첫 번째 열로 위치)
- 데이터프레임의 인덱스는 정수로 된 디폴트 인덱스로 변경됨

```python
df2.reset_index() # C1이 맨 처음 열로 이동하고
# 데이터프레임의 인덱스는 정수 인덱스(0 1 2 3 4)로 변경됨
```

```python
df2
```

reset_index() 할 때 drop=True로 설정하면,
인덱스를 열 인덱스로 올리는 것이 아니라 그냥 버림

```python
df2 = df2.reset_index(drop=True)
df2
```

```python
# 인덱스를 열 인덱스로 올리고 인덱스 순서 변경 연습하기 위해
# C1을 인덱스로 설정하고 df2 생성
df2 = df1.set_index('C1')
df2
```

```python
# C2를 인덱스 설정 : C1은 다시 첫 번째 열 인덱스로 위치
df2 = df1.set_index('C2')
df2
```

```python
# 이 상태에서 reset_index() 수행하면 C2가 인덱스 위치
df2 = df2.reset_index()
df2
```

```python
# 열 인덱스 순서 변경 : C1, C2, C3, C4
df2 = df2[['C1','C2','C3','C4']]
df2
```

### 다중 인덱스 (Multi-index)

다중 인덱스
- 행이나 열에 여러 계층을 가지는 인덱스 

열에서 다중 인덱스 생성 방법
- 데이터프레임 생성할 때
- columns 인수에 여러 개의 리스트 형태로 인덱스를 넣어서 생성

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5,4),2),
                  columns = [['A','A','B','B'],
                            ['C1','C2','C1','C2']])
df3
```

다중 인덱스에 이름 지정
- 다중 인덱스는 이름을 지정하면 편리하게 사용할 수 있음
- 열 인덱스들의 이름 지정은 columns 객체의 names 속성에 리스트로 넣어서 지정

```python
df3.columns.names = ['Cidx1','Cidx2']
df3
```

행에서 다중 인덱스 생성
- 데이터프레임 생성할 때 index 인수에 여러 개의 리스트 형태로 인덱스를 넣어서 생성
- 행 인덱스 이름 지정은 index 객체의 names 속성에 리스트로 넣어 줌

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                  columns = [['A','A','B','B'],
                             ['C','D','C','D']],
                  index = [['M','M','M','F','F','F'],
                           ['id_'+str(i+1) for i in range(3)] * 2])
df4.columns.names = ['Cidx1','Cidx2']
df4.index.names = ['Ridx1','Ridx2']
df4
```

### 행 인덱스와 열 인덱스 교환

열 인덱스를 행 인덱스로 바꾸거나 행 인덱스를 열 인덱스로 바꿀 수 있음

- stack() : 열 인덱스 -> 행 인덱스로 변환
            - 열 인덱스가 반시계 방향으로 90도 회전한 것과 비슷한 모습으로 변경

- unstack() : 행 인덱스 -> 열 인덱스로 변환
               - 행 인덱스가 시계 방향으로 90도 회전한 것과 비슷한 모양으로 변경

인덱스를 지정할 때는 문자 인덱스와 정수 인덱스 모두 사용할 수 있음

```python
# 열 인덱스를 행 인덱스로 변경 : stack() 함수 사용
# 지금의 경우는 문자 인덱스 사용
df4.stack('Cidx1')
```

```python
# 열 인덱스를 행 인덱스로 변경 : stack() 함수 사용
# 지금의 경우는 정수 인덱스 사용
# 첫 번째 열 인덱스
df4.stack(0) # == df4.stack('Cidx1')과 동일
```

```python
# Cidx2를 행 인덱스 변경 : 문자 인덱스로 사용
df4.stack('Cidx2')
```

```python
# 두 번째 열 인덱스를 행 인덱스 변경 : 정수 인덱스 사용
df4.stack(1)
```

```python
# 행 인덱스 Ridx1을 열 인덱스로 변경
df4.unstack('Ridx1')
```

```python
# 행 인덱스 Ridx2를 열 인덱스로 변경
df4.unstack('Ridx2')
```

### 다중 인덱스의 순서 교환

다중 인덱스의 순서를 바꾸고 싶으면 swaplevel() 함수 사용
- swaplevel(i, j, axis)
- i, j는 교환하고자 하는 인덱스
- axis는 0일 때는 행 인덱스, 1일 때는 열 인덱스 (디폴트 행 인덱스)

```python
# 행 인덱스 교환
df5 = df4.swaplevel('Ridx1','Ridx2') #0이 디폴트로 생략
df5
```

```python
# 열 인덱스 교환 : axis = 1
df6 = df4.swaplevel('Cidx1','Cidx2',1)
df6
```

연습 문제2
A 반 학생 5명과 B반 학생 5명의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다. 번호는 각 반마다 ID0~ID4로 지정
1. "반", "번호", "국어", "영어", "수학" 을 열로 가지는 데이터프레임 df_score3을 만든다.

2. df_score3을 변형하여 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"을 가지는 데이터프레임 df_score4을 만든다.

3. 데이터 프레임 df_score4에 각 학생의 평균을 나타내는 행을 오른쪽에 추가한다.

4. df_score3을 변형하여 행 인덱스로 "번호"를, 1차 열 인덱스로 "국어", "영어", "수학"을, 2차 열 인덱스로 "반"을 가지는 데이터프레임 df_score5을 만든다.

5. 데이터 프레임 df_score5에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.

```python
# A 반 학생 5명과 B반 학생 5명의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다. 번호는 각 반마다 ID0~ID4로 지정
np.random.seed(1)
df_score3 = pd.DataFrame({
    '반': ['A']*5+['B']*5,
    '번호' : ['ID'+str(i) for i in range(5)] * 2,
    '국어' : np.random.randint(10,101,10),
    '영어' : np.random.randint(10,101,10),
    '수학' : np.random.randint(10,101,10)},
    columns = ['반','번호','국어','영어','수학'])

df_score3
```

```python
# 2. df_score3을 변형하여 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"을 가지는 데이터프레임 df_score4을 만든다.
df_score4 = df_score3.set_index(['반','번호'])
df_score4
```

```python
# 3. 데이터 프레임 df_score4에 각 학생의 평균 값을 행 오른쪽에 추가한다.
# 평균 열 추가
df_score4.loc[:, '평균'] = df_score4.mean(axis=1).round(2)
df_score4
```

```python
# 4. df_score3을 변형하여 행 인덱스로 "번호"를, 1차 열 인덱스로 "국어", "영어", "수학"을, 
# 2차 열 인덱스로 "반"을 가지는 데이터프레임 df_score5을 만든다.
df_score5=df_score3.set_index(['번호','반']).unstack(1)
df_score5
```

```python
# 5. 데이터 프레임 df_score5에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.
df_score5.loc['반평균',:] = df_score5.mean()
df_score5
```

```python
# 연습문제3
# 아래와 같은 도시의 인구 데이터가 있다.
data = {
    "인구": [9726787,3408347,2431523,2950972,1456096,1471650,
           1144098,345216,13311254,1537780,1596613,2118457,
           1811619,1857083,2651054,3353380,670595],
    "지역": ['서울특별시','부산광역시','대구광역시','인천광역시','광주광역시',
           '대전광역시','울산광역시','세종특별자치시','경기도','강원도',
           '충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도']
  
}

# 열방향 인덱스와 행방향 인덱스 지정
columns = ["지역", "인구"]


# DataFrame(데이터, index=행방향 인덱스, columns=열방향 인덱스)
df3 = pd.DataFrame(data, columns=columns)
df3

# 1. 지역 데이터를 행 인덱스로 설정한다.
# 2. 각 지역의 인구수를 4개의 카테고리로 분류하고, 열['규모']에 해당 카테고리를 표시한다.
#카테고리 : 대,중,소,최소 - 각 구간은 아래와 같다.
# 3000000  초과 : 대
# 1000000  초과 : 중
# 500000 초과 : 소
# 그 외 : 최소   
```

```python
# 1. 지역 데이터를 행 인덱스로 설정한다.
df3 = df3.set_index('지역')
df3
```

```python
# 2. 각 지역의 인구수를 4개의 카테고리로 분류하고, 열['규모']에 해당 카테고리를 표시한다.
#카테고리 : 대,중,소,최소 - 각 구간은 아래와 같다.
# 3000000  초과 : 대
# 1000000  초과 : 중
# 500000 초과 : 소
# 그 외 : 최소   

bins = [0, 500000,1000000, 3000000, 15000000] # 경기도 인구수 천삼백만(15000000)
labels = ['최소','소','중','대']

df3['규모'] = pd.cut(df3['인구'], bins, labels=labels)
df3
```
