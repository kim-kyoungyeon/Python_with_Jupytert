```python
# 수평바 그리기 barh() 함수사용

np.random.seed(0)
people=['몽룡','춘향','방자','형단']

y= np.arange(len(people))
#array([0, 1, 2, 3])
performance= 3+10*np.random.rand(len(people))

plt.title('수평 Bar Chart')
plt.barh(y, performance, alpha= 0.4) # x위치에 ->y값 줌
plt.yticks(y,people)

plt.xlabel('퍼포먼스') 
plt.ylabel('이름')

plt.grid(True) # 수치 확인

plt.show()
```

```python
#버플 차트 점 하나의 크기 .색상이용 
#다른 데이터값 표시
# s 인수 크기
# c 인수 색상

N= 30
np.random.seed(0) #의미가? 파이썬에서 시드를 설정하는 명령은 seed이다. 인수로는 0과 같거나 큰 정수를 넣어준다.
x= np.random.rand(N) #0과 1사이의 실수 랜덤생성 0<x<1
y=np.random.rand(N) #0과 1사이의 실수 랜덤생성 0<y<1
```

```python
#색상에 적용할 변수
y2 = np.random.rand(N)
#크기에 적용할 변수
y3 = np.pi* (15*np.random.rand(N)) **2
plt.title('Bubble Chart')
plt.scatter(x,y,c=y2,s=y3) #c 색상 s크기
plt.show()
```

```python
x=np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 3*np.sin(x+1.5)

plt.figure(figsize=(10,6)) # figsize 뜻?
plt.plot(x,y1,x,y2,x,y3,x,y4) # 총4개 선 x y1 1개
plt.show()
```

# 회귀분석이란? 
-하나 또는그 이상의 독립변수 / 종속변수에 대한 영향을 추정할수있는 통계기법
- 2개의 변수에 상관관계를 기본 = 하나의 1차 선형식으로 두 변수의 관계를 일반화 하는 분석 방법
- 예를 들어 독립변수가 종속변수의 영향을 미치는가?
독립변수가 1 증가할때 종속변수는 얼마나 증가할 것인가?

회귀선에서 가장 유용한 값은 기울기
-기울기 = 종속변수에 독립변수가 미치는 영향을 나타낸다

-기울기 클수록 종속 변수에 미치는 영향, 관계의 정도가 커진다는 의미
-기울기가 - 음방향이라면 변화의 뱡향이 반대로 독립변수가 증가할 수록 종속 변수는 감소하는 관계







```python
# 카테고리 변수 smoker가 포함된 경우의 회귀선 
sns.lmplot(x= 'total_bill',y='tip', hue='smoker',
          data=tips,palette='Set1',height= 7 ) # data변수
plt.show()
```

```python
#겹치지 않고 따로 표시 # col =범주형 값
sns.lmplot(x= 'total_bill',y='tip',hue='smoker', data=tips,
          col='smoker',height=7)
plt.show()
```

#heatmap()
- x축과 y축 2개의 범주형 자료의 계급별로 연속형자료를 집계한자료를사용해서
집계한 값에 비례해서 색상을 다르게 하여 2차원으로 데이터를 시각화하는 방법

```python
# heatmap 에 사용할 데이터셋 항공가 승객수 데이터셋사용
flights= sns.load_dataset('flights')
flights.head()
```

```python
#연도별 월별 항공기 승객수 출력 ## 많이쓴다
#pivot 함수 사용해 변환 
# 행 인덱스 : 월별 
# 열 인덱스 : 연도 
#pivot( 행인덱스, 열 인덱스 ,데이터)
flights= flights.pivot ('month','year', 'passengers')
flights.head()
```

```python
# pivot 한 결과의 데이터셋을 csv 파일로 data 폴더에 저장
flights.to_csv('../data/flights.csv',encoding='utf-8')
    
```

```python
#heatmap 함수 사용한 그래프
plt.figure(figsize=(10,8))
sns.heatmap(flights)
plt.show()
```

```python
# 각 셀에 숫자를 출력 : annot= True
# fmf='f' 포맷 (실수)
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot= True, fmt= 'f')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot= True, fmt= 'f',cmap= 'RdBu') #실수
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot= True, fmt= 'd',cmap= 'RdBu') #정수
plt.show()
```

```python
# Open street Map 기반 지도
map_osm = folium.Map(location= [45.5236,-122.6750], zoom_start=13)
map_osm
```

```python
# 지도 스타일 변경 : tiles 인수
# Open Street Map 을 기반으로 동작
# 내부적으로 Stamen Terrain, Stamen Toner 두개의 스타일 지정
map_osm = folium.Map(location= [45.5236,-122.6750], zoom_start=13, tiles = 'Stamen Toner')
map_osm
```

```python
# 내부적으로 Stamen Terrain, Stamen Toner 두개의 스타일 지정
map_osm = folium.Map(location= [45.5236,-122.6750], zoom_start=13, tiles = 'Stamen Terrain')
map_osm
```

```python
map_1 = folium.Map(location= [45.372, -121.6972],
                     zoom_start=11, 
                     tiles = 'Stamen Terrain')

folium.Marker([45.3288,-121.6625],
             popup='Mt.Hood Meadowd',
             icon= folium.Icon(icon='cloud')).add_to(map_1)

folium.Marker([45.3311,-121.77135],
             popup='Timerline Lodge',
             icon= folium.Icon(color='red')).add_to(map_1)

folium.Marker([45.3300,-121.6823],
             popup='Some other location',
             icon= folium.Icon(color='green',icon= 'info-sign')).add_to(map_1)
map_1
```

```python
#서클 마커
# folium.CircleMarker(radius-크기,color-선색상, fill_color-면색상 )
# 지도에다가 부착해야한다 add_to(지도명)
map_1 = folium.Map(location= [45.3288, -121.6972],
                     zoom_start=11, 
                     tiles = 'Stamen Terrain')

folium.Marker([45.3288,-121.6625],
             popup='Mt.Hood Meadowd',
             icon= folium.Icon(icon='cloud')).add_to(map_1)
folium.CircleMarker([45.3288,-121.6625], 
                   popup='Mt.Hood Meadowd',
                   radius= 50,
                   color= 'red',
                   fill_color='blue').add_to(map_1)
map_1
```

```python
# 서울특별시청의 위도 37도 56분 63.45초이고
# 경도는 126도 97분 78.93초이다
seoul_ct_map= folium.Map(location= [37.566345, 126.977893],
                     zoom_start=17)
seoul_ct_map
```

```python
# 서울 시청 마커 표시 , popup출력 : 서울시청
# 덕수궁에 서클 마커 표시, popup 출력 : 
#덕수궁 : 37.5658049 , 126.97514610000007,
seoul_ct_map= folium.Map(location= [37.566345, 126.977893],
                     zoom_start=17)
folium.Marker([37.566345, 126.977893],
              popup='서울시청',
              icon= folium.Icon(icon='star',color='red')).add_to(seoul_ct_map)
folium.Marker([37.5658049 ,126.97514610000007],
               popup='덕수궁',
              icon= folium.Icon(icon='flag')).add_to(seoul_ct_map)
folium.CircleMarker([37.5658049 ,126.97514610000007],
                    popup='덕수궁',
                   radius= 50,
                   color= 'red',
                   fill_color='blue').add_to(map_1).add_to(seoul_ct_map)
seoul_ct_map
seoul_ct_map.save('../data/seoul.city_map.html')
```
