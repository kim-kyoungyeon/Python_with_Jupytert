경우에 따라 데이터프레임에 인덱스로 들어가 있어야 할 데이터가 일반 데이터 열에 들어가있거나 
반대로 일반 데이터 열 이여야 하는데 인덱스로 들어가있는 경우가 존재
인덱스의 열을 교환하는 방법이필요
데이터프레임의 인덱스설정하고 제거하는함수

```python
set_index() : 기존의 행 인덱스 제거 후 데이터열중에서 하나를 인덱스로 설정
    (열> 인덱스)
reset_index() :  기존의 행 인덱스ㅇ를제거하고 인덱스를 열로추가 (인덱스>열)
```

```python
import pandas as pd
import numpy as np
```

```python
# 데이터프레임생성
#np.vstack ()함수를 사용해서 데이터프레임생성
```

```python
a= np.array([1,2,3])
b= np.array([4,5,6])
```

```python
#np.vstack 함수 2개의 array를 위아래로 합쳐줌
np.vstack([a,b])
```

```python
np.hstack([a,b])# 옆으로 합침
```

```python
# np.vstakc 함수를 사용해서 데이터프레임 생성
df= pd.DataFrame(np.vstack([a,b]),
                columns = ['C1','C2','C3'])
df
```

```python
# 인덱스 조작에 사용할 데잉터프레임 생성 np.vstack 사용
#실행할때마다 동일한 값으로 실행되도록 시드값설정
np.random.seed(0)
#4행 5열에 데이터프레임생성
df1 = pd. DataFrame(np.vstack([list('ABCDE'),
                               np.round(np.random.rand(3,5),2)]),
                                index = ['a','b','c','d'])
df1
```

```python
# abcd 를 열인덱스 사용
df1.T
```

```python
df1 = pd. DataFrame(np.vstack([list('ABCDE'),
                               np.round(np.random.rand(3,5),2)]).T,
                                columns= ['c1','c2','c3','c4'])
df1
```

set_index()
 기존의 행 인덱스를 제거하고
데이터열중에서 하나를 인덱스로 설정
열에서 인덱스로 간다


```python
df2 = df1.set_index('c1')
df2
```

```python
#c2열 인덱스로 지정
#기존의 인덱스 c2은 없어짐
df2.set_index('c2')
```

```python
# reset_ index() 기존의 행인덱스를제거하고인덱스를열로추가 인덱스-> 열
# 이때인덱스열은 열이름 맨 앞에 삽입 첫번째 열로 위치
# 데이터플에미인덱스는 정수로 된 디폴트 인덱스로 변경됨
df2.reset_index() #c1이 맨 처음 열로 이동하고
#데이터 프레임
```

```python
df2
```

```python
#reset_ index()할때 drop=True로 설정하면
#인덱스를 열 인덱스로 올리는것이 아니라 그냥 버림
```

```python
df2 = df2.reset_index(drop=True)
df2
```

```python
#인덱스를 열인덱스로올리고 인덱스 순서를 변경 연습하기위해 c1을인덱스로 설정하고 df2 다시 생성
#c1 인덱스로 설정하고 df2 다시 완성
df2= df1.set_index('c1')
df2
```

```python
#c2를 인덱스 설정 c1은 첫번째  인덱스로 위치
df2= df1.set_index('c2')
df2
```

```python
# 이 상태에서 reset_index() 수행하면 c2가 인덱스 위치

df2 = df2.reset_index()
df2
```

```python
#열인덱스 순서 변경 c1 c2 c3 c4 
df2= df2[['c1','c2','c3','c4']]
df2
```

행에서 다중 인ㄷ게스 생성.
데이터프레임 생성시 index 인수에 여러개으 ㅣ리스트 형태로 인ㄷ게스를 넣어생성
형인덱스이름지정은 index 객체의 names 속성에[ 리스트 넣어줌]

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                  columns= [['a','a','b','b'],['c','d','c','d']],
                   index=[['m','m','m','f','f','f'],['id_'+str(i+1) for i in range(3)]*2])


df4.columns.names=['cidx1','cidx2']

df4.index.names=['ridx1','ridx2']
df4
```

# 행인덱스와 열 인덱스와 교환 
열 인덱스 를 행인ㄷ게스로 바꾸거나 서로서로 바꿀수있따

-stack () 열 인덱스> 행인덱스 로 변환
    열 인덱스가 반시계 방향으로 90도 회전한 것과 비슷한 모습으로 변경된다
unstack() 행인덱스>열인덱스변환
    행 인덱스가 시계방향으로 90도 회전한것과 비슷한 모양변경
    
인덱스를 지정할때는 문자인덱스와 정수인덱스모두사용할수잇음

```python
#열인덱스를 행인데긋로 변경 : stack() 함수사용
#문자인덱스 사용 
df4.stack('cidx1')
```

```python
#열인덱스를 행인데긋로 변경 : stack() 함수사용
#정수인덱스 사용 
#첫번째 열인덱스
df4.stack(0) # df4.stack('cidx1')  동일
```

```python
#cidx2 행인덱스로 변경 :문자인덱스사용
df4.stack('cidx2') 
```

```python
#두번째 열인덱스 행인덱스변경 정수인덱스사용
df4.stack(1) 
```

```python
#행인데긋 ridx1을 열인덱스로 변경
df4.unstack('ridx1')
```

```python
#행인데긋 ridx2을 열인덱스로 변경
df4.unstack('ridx2')
```

```python
#첫번째 행 인ㄷ게스를 열인ㄷ게스로 변경
df4.unstack(0)
```

```python
#두번째 행 인ㄷ게스를 열인ㄷ게스로 변경
df4.unstack(1)
```

```python
df4.sum(axis=0) #각열의합계
```

```python
df4.sum(axis=1) #각행의합계
```

```python
# All 행 인덱스 추가 : 각 열의 합
df4.loc[('All','All'),:] = df4.sum()  #디폴트 axis =0
df4
```

```python
# All 행 인덱스 추가 : 각 행의 합
# RowSum 이라는 열추가 : 값은 각 행의 합계
df4.loc[:,'RowSum'] = df4.sum(axis=1)
df4
```

```python
## 다중 인덱스의 순서 교환
# 다중 인덱스의 순서를 바꾸고 싶으면 swaplevel() 함수 사용
# - swaplevel(i,j,axis)
# - i,j는 교환하고자 하는 인덱스 
# axis는 0일때는 행 인덱스, 1일때는 열 인덱스 (디폴트는 행인덱스)

#행인덱스 교환
df5 = df4.swaplevel('ridx1','ridx2') #0이 디폴트로생략
df5
```

```python
#열 인덱스 교환 axis 1
df6 = df4.swaplevel ('cidx1','cidx2',1)
df6
```

```python
#다중인덱스가 있느 경우에 정렬
# 다중 인덱스가 있는 데이터프레임을 sort_index()로 정렬할떄
# level 인수를 사용해서 
#어떤 인덱스를 기준으로 정렬할지를 알려 줘야함 

df5
```

```python
df5.sort_index(level=0)
```

```python
df6
```

```python
df6.sort_index(axis=1,level=0)
```

```python
# 연습 문제2
# A 반 학생 5명과 B반 학생 5명의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 다음과 같이 만든다. 번호는 각 반마다 ID0~ID4로 지정
# 1. "반", "번호", "국어", "영어", "수학" 을 열로 가지는 데이터프레임 df_score3을 만든다.

df_score3= pd.DataFrame ({
    "반":['A']*5 +['B']*5,
    "번호":['ID'+str(i) for i in range (5)]*2,
    "국어":np.random.randint(0,101,10),
    "영어":np.random.randint(0,101,10),
    "수학":np.random.randint(0,101,10)}, # 최소값 최대값 10개 
    columns=["반", "번호", "국어", "영어", "수학"])
df_score3


```

```python
# 2. df_score3을 변형하여 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"을 가지는 데이터프레임 df_score4을 만든다.
df_score4=df_score3.set_index(['반','번호'])
df_score4


```

```python
# 3. 데이터 프레임 df_score4에 각 학생의 평균을 나타내는 행을 오른쪽에 추가한다.
#평균열추가
df_score4.loc[:,'평균'] =df_score4.mean(axis=1).round(2)
df_score4

```

```python
## 4. df_score3을 변형하여 행 인덱스로 "번호"를, 1차 열 인덱스로 "국어", "영어", "수학"을, 2차 열 인덱스로 "반"을 가지는 데이터프레임 df_score5을 만든다.
df_score5 = df_score3.set_index(['번호', '반']) .unstack(1)
df_score5
```

```python

# 5. 데이터 프레임 df_score5에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.
df_score5.loc['반평균',:] =df_score5.mean()
df_score5
```

```python
# 연습문제3
# 아래와 같은 도시의 인구 데이터가 있다.
data = {
    "인구": [9726787,3408347,2431523,2950972,1456096,
    1471650,1144098,345216,13311254,1537780,
    1596613,2118457,1811619,1857083,
    2651054,3353380,670595],
    "지역": ['서울특별시','부산광역시','대구광역시','인천광역시','광주광역시',
    '대전광역시','울산광역시','세종특별자치시','경기도',
    '강원도','충청북도','충청남도','전라북도','전라남도',
    '경상북도','경상남도','제주특별자치도']
  
}

# 열방향 인덱스와 행방향 인덱스 지정
columns = ["지역", "인구"]



# DataFrame(데이터, index=행방향 인덱스, columns=열방향 인덱스)
df3 = pd.DataFrame(data, columns=columns)
df3
```

```python
# 2. 각 지역의 인구수를 4개의 카테고리로 분류하고, 열['규모']에 해당 카테고리를 표시한다.
# #카테고리 : 대,중,소,최소 - 각 구간은 아래와 같다.
# # 3000000  초과 : 대
# # 1000000  초과 : 중
# # 500000 초과 : 소
# # 그 외 : 최소  
bins = [0,500000,1000000,5000000,15000000]
labels = ['최소','소','중','대']
df3['규모'] = pd.cut(df3['인구'],bins, labels=labels)
df3
```
