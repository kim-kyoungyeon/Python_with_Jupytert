시리즈부터 시작
- 시리즈는 데이터가 순차적으로 나열된 1차원 배열형태
- 인덱스 index 와 data 값 (value) 이 1:1로 대응
-딕셔너리와 비슷한 구조 : [key:value]
-시리즈의 인덱스는  = 데이터의 값의 위치를 나타내는 이름표 (숫자,문자)

```python
#정답
data= {
     '국어':[80,90,100,20,10],
     '영어':[70,80,90,100,20],
     '수학':[100,20,80,50,10],
     '음악':[10,20,80,90,100],
     '체육':[50,70,80,10,90]
}

index=['손흥민', '김연아','박지성','손연재','박병호']
df= pd.DataFrame(data,index=index)
df

#(2) 영어 점수 변경 : 95,87,75,80,98
df['영어']= [95,87,75,80,98]
df
#(3) 열삭제 :'음악' '체육' 삭제
df=df.drop(['음악','체육'],axis=1)
df
#(4) 열추가 : '총점 평균
df['총점'] =df['국어']+df['영어']+df['수학']
df
df['평균'] = round((df['총점']/3),2)
df
#(5) 행삭제 ; 박지성
df = df.drop('박지성')
df
#(6) 행추가 임의로 3개행 추가
new_rows= [
    [90,90,90,270,90],
    [80,80,80,240,80],
    [70,70,70,210,70]    
]
new_df = pd.DataFrame (new_rows, columns= df.columns,
                      index= ['홍길동','이몽룡','성춘향'])
new_df
# 하나의 셀에 다 넣으면안됨
```
