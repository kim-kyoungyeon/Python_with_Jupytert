```python
import seaborn as sns
import numpy as np
import pandas as pd 
titanic= sns.load_dataset('titanic')
```

1. qcut 명령으로 세 개의 나이 그룹을 만든다.¶
2.성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.¶
3.성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.¶

```python
titanic= sns.load_dataset('titanic')
titanic.tail(200)
```

```python
titanic["나이"] = pd.qcut(titanic["age"], 3, labels=["1그룹 : 25세 이하","2그룹 : 25~35세 이하","3그룹 : 35세 이상"]) # qcut 과 pivot의 비교  # 
titanic.tail(200)
```

```python
qcut 은 동일개수로 나누어서 범주만들기! / 일반 cut은 동일한 길이로 나누어서 범주만든다
```

```python
titanic.groupby(["sex", "나이", "class"])[['survived']].sum()  / titanic.groupby(["sex", "나이", "class"])[['survived']].count() # 선별  group+by [추출1,추출2추출열3] [[]].sum(새로운 변수) 
# sum/ count 평균 계산한거임.
```

```python
pivotdata = titanic.pivot_table('survived', index=['sex', '나이'],aggfunc='mean', columns='class') # pivot_table('기준열', index= [인덱싱할열, 열2] aggfunc= 산수함수 , columns= 열이름)


pivotdata
```
